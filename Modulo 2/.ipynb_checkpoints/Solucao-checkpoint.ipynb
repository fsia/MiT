```python
# Estes são os pacotes necessários para esta atividade
import pandas as pd
import numpy as np

# Use o Pandas para ler o arquivo csv em um dataframe
# Note que o delimitador neste csv é o ponto e vírgula ";" em vez de uma vírgula ","
df_red = pd.read_csv('winequality-red.csv',delimiter=";")

# Como estamos realizando uma tarefa de classificação, vamos atribuir a todos os vinhos tintos uma etiqueta de 1
df_red["color"] = 1

# O método .head() é útil para visualizar uma prévia dos nossos dados.
df_red.head()
```

```python
df_white = pd.read_csv('winequality-white.csv',delimiter=";")
df_white["color"] = 0  # Atribuir vinho branco ao rótulo 0
df_white.head()
```

```python
# Agora combinamos nossos dois dataframes:
df = pd.concat([df_red, df_white])

# Embaralhamos eles no local para misturar os dados de vinho tinto e vinho branco juntos
df = df.sample(frac=1).reset_index(drop=True)
df.head()
```

```python
# Escolhemos três atributos do vinho para realizar nossa previsão
input_columns = ["citric acid", "residual sugar", "total sulfur dioxide"]
output_columns = ["color"]

# Extraímos os recursos relevantes para nossas matrizes (arrays) numpy X e Y:
X = df[input_columns].to_numpy()
Y = df[output_columns].to_numpy()
print("Shape of X:", X.shape)
print("Shape of Y:", Y.shape)
in_features = X.shape[1]
```

```python
## SOLUÇÃO

# Aqui, copiamos a implementação da nossa classe base SingleNeuronModel:
class SingleNeuronModel():
    def __init__(self, in_features):
        # Definimos os pesos iniciais com valores pequenos normalmente distribuídos
        self.w = 0.01 * np.random.randn(in_features)
        self.w_0 = 0.01 * np.random.randn()
        self.non_zero_tolerance = 1e-8 # Adicionar isso para evitar divisões por zero

    def forward(self, x):
        # Calcular e salvar a pré-ativação z
        self.z = x @ self.w.T + self.w_0


        # Aplicar a função de ativação e retornar
        self.a = self.activation(self.z)
        return self.a

    def activation(self, z):
        raise ImplementationError("o método de ativação deve ser implementado pela subclasse")

    # Calcular e salvar o gradiente de saída em relação aos pesos
    def gradient(self, x):
        raise ImplementationError("o método de gradiente deve ser implementado pela subclasse")

    # Atualizar os pesos com base nos gradientes e na taxa de aprendizado
    def update(self, grad_loss, learning_rate):
        self.w   -= grad_loss * self.grad_w   * learning_rate
        self.w_0 -= grad_loss * self.grad_w_0 * learning_rate


# Modelo de classificação de neurônio único
class SingleNeuronClassificationModel(SingleNeuronModel):
    # Função de ativação sigmoidal para classificação
    def activation(self, z):
        return 1 / (1 + np.exp(-z) + self.non_zero_tolerance)

    # Gradiente da saída em relação aos pesos, para ativação sigmoidal
    def gradient(self, x):
        self.grad_w = self.a * (1-self.a) * x
        self.grad_w_0 = self.a * (1-self.a)

# Processo de treinamento, utilizando perda de logaritmo negativo da verossimilhança (NLL) -
# apropriada para problemas de classificação
def train_model_NLL_loss(model, input_data, output_data,
                         learning_rate, num_epochs):
    non_zero_tolerance = 1e-8 # Adicionar isso para evitar cálculos de log de 0
    num_samples = len(input_data)
    for epoch in range(1, num_epochs+1):
        total_loss = 0 # Monitorar a perda total em todo o conjunto de dados

        for i in range(num_samples):
            x = input_data[i,...]
            y = output_data[i]
            y_predicted = model.forward(x)
            # Função de perda de NLL
            loss = -(y * np.log(y_predicted + non_zero_tolerance) + (1-y) * np.log(1-y_predicted + non_zero_tolerance))
            total_loss += loss

            # Gradiente da previsão em relação aos pesos
            model.gradient(x)

            # Gradiente da perda em relação à previsão, para NLL
            grad_loss = (y_predicted - y)/(y_predicted * (1-y_predicted))

            # Atualizar nosso modelo com base nos gradientes
            model.update(grad_loss, learning_rate)

        report_every = max(1, num_epochs // 10)
        if epoch == 1 or epoch % report_every == 0: # A cada poucas épocas, relatar
            print("epoch", epoch, "has total loss", total_loss)

# Usaremos essa função para avaliar o desempenho do nosso classificador treinado
def evaluate_classification_accuracy(model, input_data, labels):
    # Contar o número de amostras classificadas corretamente dado um conjunto de pesos
    correct = 0
    num_samples = len(input_data)
    for i in range(num_samples):
        x = input_data[i,...]
        y = labels[i]
        y_predicted = model.forward(x)
        label_predicted = 1 if y_predicted > 0.5 else 0
        if label_predicted == y:
            correct += 1
    accuracy = correct / num_samples
    print("Our model predicted", correct, "out of", num_samples,
          "correctly for", accuracy*100, "% accuracy")
    return accuracy
```

```python
# Criar um modelo SingleNeuronClassification que pode prever com base em nossas 3 características escolhidas.
classification_model = SingleNeuronClassificationModel(in_features)

# Treinar o modelo em nosso conjunto de dados. Aviso: isso levará alguns minutos
learning_rate = 0.001
epochs = 200
train_model_NLL_loss(classification_model, X, Y, learning_rate, epochs)

# Verificar a precisão de classificação final do nosso modelo
evaluate_classification_accuracy(classification_model, X, Y)
```

Conseguimos alcançar cerca de 91% usando nosso modelo. É um desempenho muito bom e certamente poderia ser melhorado incluindo mais recursos e utilizando uma arquitetura de rede neural mais complexa.
